```python
#convert image from bgr to rgb
img=cv2.imread('aulogo.png')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
#convert image from rgb to bgr
img=cv2.imread('aulogo.png')
img=cv2.cvtColor(img,cv2.COLOR_RGB2BGR)
plt.imshow(img)
```
