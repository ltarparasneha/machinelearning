```python
#box blur
import cv2
import matplotlib.pyplot as plt
import numpy as np
img=cv2.imread('cloud.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
box=cv2.blur(img,(10,10))
plt.subplot(1,2,1),plt.imshow(img),plt.axis('off')
plt.subplot(1,2,2),plt.imshow(box),plt.axis('off')
```

```python
#gaussian blur
import cv2
import matplotlib.pyplot as plt
import numpy as np
img=cv2.imread('mind.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
gaus=cv2.GaussianBlur(img,(15,15),12)
plt.subplot(1,2,1),plt.imshow(img),plt.axis('off')
plt.subplot(1,2,2),plt.imshow(gaus),plt.axis('off')
```

```python
#median blur
import cv2
import matplotlib.pyplot as plt
import numpy as np
img=cv2.imread('landscape.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
med=cv2.medianBlur(img,51)
plt.subplot(1,2,1),plt.imshow(img),plt.axis('off')
plt.subplot(1,2,2),plt.imshow(med),plt.axis('off')
```

```python
#box blur
import cv2
import matplotlib.pyplot as plt
import numpy as np
img=cv2.imread('flower1.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
box=cv2.blur(img,(30,30))
gaus=cv2.GaussianBlur(img,(15,15),5)
med=cv2.medianBlur(img,15)
plt.subplot(2,2,1),plt.imshow(img),plt.title('source image'),plt.axis('off')
plt.subplot(2,2,2),plt.imshow(box),plt.title('boxblur image'),plt.axis('off')
plt.subplot(2,2,3),plt.imshow(gaus),plt.title('gaussian image'),plt.axis('off')
plt.subplot(2,2,4),plt.imshow(med),plt.title('median image'),plt.axis('off')
```

```python
#motion blur
#(i)horizontal blur
import cv2
import matplotlib.pyplot as plt
import numpy as np
img=cv2.imread('tree.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
size=30

k1=np.zeros((size,size))
k1[int(size/2),:]=np.ones(size)
k1=k1/size

k2=np.zeros((size,size))
k2[:,int(size/2)]=np.ones(size)
k2=k2/size

horizontal=cv2.filter2D(img,-1,k1)
vertical=cv2.filter2D(img,-1,k2)

plt.subplot(1,3,1),plt.imshow(img),plt.title('source image'),plt.axis('off')
plt.subplot(1,3,2),plt.imshow(horizontal),plt.title('horizotal image'),plt.axis('off')
plt.subplot(1,3,3),plt.imshow(vertical),plt.title('vertical image'),plt.axis('off')
```
