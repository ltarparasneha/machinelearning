```python
#bilateral
import cv2
import matplotlib.pyplot as plt

img=cv2.imread('rose.jpg')
bilateral=cv2.bilateralFilter(img,d=9,sigmaColor=75,sigmaSpace=75)

plt.subplot(1,2,1),plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB)),plt.title('source'),plt.axis('off')
plt.subplot(1,2,2),plt.imshow(cv2.cvtColor(bilateral,cv2.COLOR_BGR2RGB)),plt.title('bilateral'),plt.axis('off')
```

```python
#unsharp
import cv2
import matplotlib.pyplot as plt

img=cv2.imread('aerial.png')

gauss=cv2.GaussianBlur(img,(9,9),0)
unsharp=cv2.addWeighted(img,1.5,gauss,-0.5,0)

plt.subplot(1,2,1),plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB)),plt.title('source'),plt.axis('off')
plt.subplot(1,2,2),plt.imshow(cv2.cvtColor(unsharp,cv2.COLOR_BGR2RGB)),plt.title('unsharp'),plt.axis('off')
```

```python
#scharr
import cv2
import matplotlib.pyplot as plt
img=cv2.imread('zebra.jpg',0)

scharrx=cv2.Scharr(img,cv2.CV_64F,1,0)
scharry=cv2.Scharr(img,cv2.CV_64F,0,1)

shcarrx=cv2.convertScaleAbs(scharrx)
shcarry=cv2.convertScaleAbs(scharry)

scharr_combine=cv2.addWeighted(scharrx,0.10,scharry,0.2,0)

plt.subplot(1,3,1),plt.imshow(scharrx,cmap='gray'),plt.title('scharrx'),plt.axis('off')
plt.subplot(1,3,2),plt.imshow(scharry,cmap='gray'),plt.title('scharry'),plt.axis('off')
plt.subplot(1,3,3),plt.imshow(scharr_combine,cmap='gray'),plt.title('scharr_combine'),plt.axis('off')
```
